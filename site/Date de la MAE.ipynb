```python
import json 
import pprint

pp = pprint.PrettyPrinter(indent=4)

with open('static/data.json') as json_file:
    data = json.load(json_file) 
    locations = data['locations'] 
locations = sorted(locations, key=lambda l: l['country']) 
locations_by_country = {} 
for l in locations: 
    country = l['country'] 
    address = [s.strip() for s in l['address'].split(',')] 
    coordinates = l['coordinates'] 
    name = l['name'] 
    locations_by_country.setdefault(country, []).append({
        'coordinates': coordinates, 
        'address': address, 
        'name': name
    })

for country in locations_by_country:
    for location in locations_by_country[country]:
        print(country)
        pp.pprint(location)
        print()
        print()
```
